## SEIR Model
The SIR model is often good for basic simulations, but one immediate issue crops up. A person is not always immediately infectious as soon as they come in contact with an infected person. The period in between a susceptible person coming in contact with an infected person, and that person becoming infectious, is called as the period in which this individual is **exposed**. This entails the addition of another compartment, E or exposed, in addition to the S, I and R compartments.<br>

The compartment model can be diagrammed as follows.

$$\text{Susceptible}
\xrightarrow{\frac{\beta S I}{N}} 
\text{Exposed} 
\xrightarrow{\alpha E} 
\text{Infectious} 
\xrightarrow{\gamma I} 
\text{Recovered} $$

* $\beta$ is defined as in the SIR model, the only difference being that the population now moves into the exposed compartment instead of the infected compartment.
* $\alpha E$ is the rate at which exposed population becomes infective, where $E$ is the size of the exposed population. The average period of time in the exposed state is the incubation period of the disease, and equal to $\frac{1}{\alpha}$.
* The definition of $\gamma$ is similar to that of the SIR model.

An elementary model for the spread of an infectious disease in a uniform population is given by the deterministic SEIR equations.

\begin{align*}
\frac{dS}{dt} & = \frac{-\beta S I}{N} \\
\frac{dE}{dt} & = \frac{\beta S I} {N} - \alpha E \\
\frac{dI}{dt} & = \alpha E  - \gamma I \\
\frac{dR}{dt} & = \gamma I
\end{align*} 

where $S + E + I + R = N$ is an invariant.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
COLORS = {"S": "#2f4b7c",
          "E": "#ffa600",
          "I": "#f95d6a",
          "R": "#a05195"}
# parameter values
R0 = 8
t_incubation = 6
t_infective = 10
alpha = 1/t_incubation
gamma = 1/t_infective
beta = R0*gamma

# initial number of infected and recovered individuals
N=1000
E_initial = 0.01 * N
I_initial = 0
R_initial = 0
S_initial = N - E_initial - I_initial - R_initial


# SEIR model differential equations.
def SEIR(x, t, N, alpha, beta, gamma):
    S, E, I, R = x
    dSdt = -beta * S * I / N
    dEdt =  beta * S * I / N - alpha * E
    dIdt = alpha * E - gamma * I
    dRdt =  gamma * I
    return [dSdt, dEdt, dIdt, dRdt]

t = np.linspace(0, 100, 100)
x_initial = S_initial, E_initial, I_initial, R_initial
soln = odeint(SEIR, x_initial, t, args=(N, alpha, beta, gamma))
S, E, I, R = soln.T

#plotting the data
def plotdata(t, S, I, E, R): 
    fig = plt.figure(figsize=(18,12))
    ax = [fig.add_subplot(221, axisbelow=True), 
          fig.add_subplot(223)]

    ax[0].plot(t, S, color = COLORS['S'], lw=3, label='Susceptible')
    ax[0].fill_between(t, 0, S, color=COLORS["S"], alpha=0.15)
    ax[0].plot(t, I, color = COLORS['I'], lw=3, label='Infectious')
    ax[0].fill_between(t, 0, I, color=COLORS["I"], alpha=0.15)
    ax[0].plot(t, R, color = COLORS['R'], lw=3, label='Recovered')
    ax[0].fill_between(t, 0, R, color=COLORS["R"], alpha=0.15)
    ax[0].plot(t, E, color = COLORS['E'], lw=3, label='Exposed')
    ax[0].fill_between(t, 0, E, color=COLORS["E"], alpha=0.15)
    ax[0].set_title('Susceptible, Exposed, Infectious and Recovered Population')
    ax[0].set_xlabel('Time in days')
    ax[0].set_ylabel('Population')

    ax[1].plot(t, I, color = COLORS['I'], lw=3, label='Infectious')
    ax[1].fill_between(t, 0, I, color=COLORS["I"], alpha=0.15)
    ax[1].plot(t, E, color = COLORS['E'], lw=3, label='Exposed')
    ax[1].fill_between(t, 0, E, color=COLORS["E"], alpha=0.15)
    ax[1].set_title('Infectious vs Exposed Population')
    ax[1].set_xlabel('Time in days')
    ax[1].set_ylabel('Population')
    
    for a in ax: 
        a.grid(True)
        a.legend()

    plt.tight_layout()

plotdata(t, S, I, E, R)
```

The addition of an exposed population compartment slows the outbreak, but doesn't appear to significantly reduce the number of people ultimately infected by the disease. 

## Mitigations

For modelling, we introduce a social distancing parameter $u$ indicating its effectiveness. $u=0$ corresponds to no distancing and $u=1$ corresponds to perfect isolation of infective individuals. The purpose of this modification is to explore how a social distancing stragtegy affects the outcome of an epidemic.

* **Exposed.** The compartment that has been exposed to the disease but not yet infective. - This is where the social distancing parameter affects things.

The compartment model can be diagrammed as follows.

$$\text{Susceptible}
\xrightarrow{(1-u)\frac{\beta S I}{N}} 
\text{Exposed} 
\xrightarrow{\alpha E} 
\text{Infectious} 
\xrightarrow{\gamma I} 
\text{Recovered} $$

For an initial observation, I will take $u=0.5$.

```python
u=0.5
def SEIR_M(x, t, N, alpha, beta, gamma):
    S, E, I, R = x
    dSdt = -(1 - u) * beta * S * I / N
    dEdt =  (1 - u) * beta * S * I / N - alpha * E
    dIdt = alpha * E - gamma * I
    dRdt =  gamma * I
    return [dSdt, dEdt, dIdt, dRdt]

t = np.linspace(0, 100, 100)
x_initial = S_initial, E_initial, I_initial, R_initial
soln = odeint(SEIR_M, x_initial, t, args=(N, alpha, beta, gamma))
S, E, I, R = soln.T

plotdata(t, S, I, E, R)
```

Social distancing has several beneficial effects:

* Slows down the progress of the epidemic.
* Reduces the fraction of the population infected at any point in time, thereby reducing strain on health care resources.
* Reduces the number of individuals ultimately experiencing the disease. For a disease with a non-zero mortaility, this saves lives.<br>

With just u = 0.5, the number of exposed individuals reduces to almost half, and the infectious individuals also reduces significantly. One immediate issue we see, though, is that the pandemic will take longer to run its course - naturally, since the socially distanced individuals are not yet exposed to the pathogen by contact, they will open themselves up to infection if they decide to stop distancing. That won't be an issue if there are no infectious individuals, but it's really unlikely. <br>In this case one might think that it might be a good idea to gather up all the infectious and unaffected individuals together, much like the idea of **'chickenpox parties'** that are held in some localities to expose the unaffected children to the affected ones so that they can also catch the infection, sooner than later and develop immunity post infection. (It must be mentioned, though, that experts advise against it considering different individuals have varied effects to the same infection.)
